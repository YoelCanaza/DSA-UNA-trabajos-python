```python
import time
import matplotlib.pyplot as plt

def iterative_version(n):
  sum = 0

  for i in range(1, n + 1):
    for j in range(1, i + 1):
      for k in range(1, j + 1):
        sum += 1

  return sum

def formula_version(n):
  return (n*(n+1)*(n+2)) / 6

def medir_tiempo_iterativo(n):
  start_time = time.perf_counter()
  iterative_version(n)
  end_time = time.perf_counter()
  return end_time - start_time

def medir_tiempo_formula(n):
  start_time = time.perf_counter()
  formula_version(n)
  end_time = time.perf_counter()
  return end_time - start_time


n_values = list(range(1, 401))

execution_times_iterative = [medir_tiempo_iterativo(n) for n in n_values]
execution_times_formula = [medir_tiempo_formula(n) for n in n_values]

results_iterative = [iterative_version(n) for n in n_values]
results_formula = [formula_version(n) for n in n_values]

print(results_iterative)
print(results_formula)

```

```python
# Imprimiendo los tiempos de ejecución de la versión iterativa
plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times_iterative, linestyle='-', color='b')
plt.title('Tiempo de Ejecución de la Versión Iterativa')
plt.xlabel('Término enésimo (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.grid(True)
plt.show()


print(iterative_version(4))
print(formula_version(4))
```

```python
# Imprimiendo los tiempos de ejecución de la versión fórmula

plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times_formula, linestyle='-', color='r')
plt.title('Tiempo de Ejecución de la Versión Fórmula')
plt.xlabel('Término enésimo (n)')
plt.ylabel('Tiempo de ejecución (segundos)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(n_values, execution_times_iterative, label='Iterative', color='b', linestyle='-')
plt.plot(n_values, execution_times_formula, label='Formula', color='r', linestyle='--')

plt.xlabel('n')
plt.ylabel('Tiempo de Ejecución (segundos)')
plt.title('Tiempo de Ejecución de Versiones Iterativa y Fórmula')

plt.legend()
plt.grid(True)
plt.show()
```
